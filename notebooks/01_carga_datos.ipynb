```python
df = pd.read_csv('../data/Cantidad_anual_de_suicidios_reportados_en_el_Departamento_de_Antioquia_desde_2005_20260503.csv')
df.info()
```

```python
columnas_a_usar = [
    'NombreMunicipio',
    'NombreRegion',
    'Anio',
    'CausaMortalidad',
    'TipoPoblacionObjetivo',
    'NumeroPoblacionObjetivo',
    'NumeroCasos'
]
```

```python
df_data = df[columnas_a_usar].copy()
df_data.head()
```

## Conversion de datos

```python
cols_categoricas = ['NombreMunicipio','NombreRegion']
df_data[cols_categoricas] = df_data[cols_categoricas].astype('category')
df_data.info()
```

## Creacion de tablas 

```python
pivot_table = pd.pivot_table(df_data, values='NumeroCasos', index='NombreMunicipio',columns='Anio', aggfunc='sum')
print(pivot_table)
```

```python
group_table_casos = df_data.groupby('Anio')['NumeroCasos'].sum()
print(group_table_casos)
```

## Figuras y Graficos

```python
fig = px.bar(df_data , x='Anio', y='NumeroCasos', title='Numero casos vs anios')
fig.show()
```
