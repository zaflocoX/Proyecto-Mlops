## **Carga de los datos**

```python
import pandas as pd
from pathlib import Path
```

```python
df = pd.read_csv('../data/Cantidad_anual_de_suicidios_reportados_en_el_Departamento_de_Antioquia_desde_2005_20260503.csv')
df.info()
```

```python
columnas_a_usar = [
    'NombreMunicipio',
    'NombreRegion',
    'Anio',
    'CausaMortalidad',
    'TipoPoblacionObjetivo',
    'NumeroPoblacionObjetivo',
    'NumeroCasos'
]
```

```python
df_data = df[columnas_a_usar]
df_data.info()
```

```python
ruta_datos = Path.cwd().resolve().parent / 'data'
ruta_datos
```

```python
ruta_archivo = ruta_datos / "01_data_suicidios.parquet"

ruta_datos.mkdir(parents=True, exist_ok=True)

df_data.to_parquet(ruta_archivo, index=False)
```
